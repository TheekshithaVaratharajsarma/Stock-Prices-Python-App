```python
import pandas as pd

file_paths = [
    "Amazon.csv",
    "Apple.csv",
    "Facebook.csv",
    "Google.csv",
    "Microsoft.csv",
    "Netflix.csv",
    "Tesla.csv",
    "Uber.csv",
    "Walmart.csv",
    "Zoom.csv"
]

dfs = []

for file in file_paths:
    company_name = file.split(".")[0]
    df = pd.read_csv(file)
    df['Company'] = company_name 
    dfs.append(df)

stock = pd.concat(dfs, ignore_index=True)

stock.to_excel("merged_data.xlsx", index=False)
```
